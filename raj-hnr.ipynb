```python
import pandas as pd
import readline, rlcompleter
import matplotlib.pyplot as plt
import sys 
import numpy as np
%matplotlib inline
pd.options.display.max_columns = 50
pd.options.display.max_rows = 200
from  sdipylib.url import cache_url
# If either the import or the assert fails, you need to run: sudo pip install -U sdipylib
import sdipylib
assert sdipylib.__version__ >= 1005
from pandas.util.testing import assert_frame_equal
```

Read the data to a data frame

```python
fn = cache_url('http://devtest.sandiegodata.org.s3.amazonaws.com/manifests/7a007ac7-960f-468b-8ce3-7b9f83e34ea8/hnr0914.csv')
df = pd.read_csv(fn)
```

Filter the dataframe to only contain cases from 2009-2012

```python
df = df[(df.complaint_year == 2009) |
        (df.complaint_year == 2010) |
        (df.complaint_year == 2011) |
        (df.complaint_year == 2012) 
        ]
```

Remove all cases where the final disposition has no value

```python
df = df[~pd.isnull(df['final_disposition'])] 
```

Remove all cases which were dismissed

```python
df = df[~df.final_disposition.str.contains('Dismissed')]
```

### Answer to part 1:

#### All the cases between 2009 and 2012 for which the final sentence (highest_charge) is:                   PC187, PC191, PC192, or VC20001

Create a subset of the original data where the higest charge is one of the above.

```python
hcDF = df[(df.highest_charge == 'PC187(a)') |
        (df.highest_charge == 'VC20001(a)') | 
        (df.highest_charge == 'VC20001(b)(2)') | 
        (df.highest_charge == 'PC191.5(a)') |
        (df.highest_charge == 'PC192(c)(1)') | 
        (df.highest_charge == 'PC191.5(b)') | 
        (df.highest_charge == 'PC187(a) - PC664')] 
```

```python
incar = df20001[(~pd.isnull(df20001['incar']))].groupby(['complaint_year','highest_charge','highest_dispo_charge']).number.count()
not_incar = df20001[(pd.isnull(df20001['incar']))].groupby(['complaint_year','highest_charge','highest_dispo_charge']).number.count()
incar_time = np.round(df20001[(~pd.isnull(df20001['incar']))].groupby(['complaint_year','highest_charge','highest_dispo_charge']).incar.mean(),2)
summary20001 = pd.DataFrame({"Incarcerated":pd.Series(incar),
                        "Not Incarcerated":pd.Series(not_incar),
                        "Average time imprisoned": pd.Series(incar_time)
                        })
summary20001 = summary20001[["Incarcerated","Not Incarcerated","Average time imprisoned"]]
summary20001 = summary20001.fillna(0)
summary20001
```

# Analysis:

From 2009 to 2012, there were 17 cases where the initial highest charge was manslaughter or murder. Almost half (8) of these cases happened in 2009. There was one case in 2011 where the convict wasn't incarcerated after being charged for vehicular manslaughter. All other convicts served time ranging from 4 years to 26 years in prison.



Most of the cases were charged with section VC20001 - Faliure to stop at an accident resulting in death or injury.
In 2009, out of the 52 cases charged with VC 20001, only 28 served any time in prison(5 out of 10 were imprisoned).
In 2010, out of the 60 cases charged with VC 20001, only 32 served any time in prison(5 out of 10 were imprisoned).
In 2011, out of the 52 cases charged with VC 20001, only 26 served any time in prison(5 out of 10 were imprisoned).
In 2012, out of the 45 cases charged with VC 20001, only 28 served any time in prison(6 out of 10 were imprisoned).
So we see that in the past 4 years any person who was caught for fleeing the site of accident which resulted in death
or injury of a person had a 50% chance of not serving time in prison. This trend has been striking constant over the last 4 years.

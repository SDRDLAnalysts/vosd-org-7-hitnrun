### Here is a quick look at the dataset

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import pandas as pd
pd.options.display.max_columns = 50

df = pd.read_csv('hnr0914.csv')

df.head()
```

```python
print "The columns that we have data for\n"
df.columns
```

# 2009 - 2012 Cases

```python
data0912 = df[(df.complaint_year >= 2009) & (df.complaint_year <= 2012)]
print "The total number of cases from 2009 to 2012 is " + str(len(data0912))
print "Here are the highest charges codes that we are looking at and what they correspond with."
print "Also included is the number of incidents between 2009 and 2012 where the highest charge is any of the below"
print "\n"
print """VC20001\t\t\t\tFailing to Stop after an Accident (Injury or Death)
PC191.5(a)\t\t\tVehicular Manslaughter (while intoxicated)
PC187(a)\t\t\tMurder
PC192(c)(1)\t\t\tVehicular Manslaughter
PC187(a) - PC664\t\tMurder - Attempted Murder \n"""
data0912subset = data0912[ (data0912.highest_charge == 'PC187(a)') | (data0912.highest_charge == 'PC187(a) - PC664') |
                    (data0912.highest_charge == 'PC191.5(a)') | (data0912.highest_charge == 'PC191.5(b)') |
                    (data0912.highest_charge == 'PC192(c)(1)') |
                    (data0912.highest_charge == 'VC20001(a)') | (data0912.highest_charge == 'VC20001(b)(2)') ]
print "Our reduced dataset only includes cases from 2009-2012 (inclusive) and have the above as the highest charges and contains " + str(len(data0912subset)) + " cases"
#data0912.complaint_year.value_counts()
data0912subset.highest_charge.value_counts()
```

# Incarcerated vs Not Incarcerated

```python
year09 = data0912subset[(data0912subset.complaint_year == 2009)]
num09incarc = year09.incar.count()
num09NotIncar = len(year09) - num09incarc

year10 = data0912subset[(data0912subset.complaint_year == 2010)]
num10incarc = year10.incar.count()
num10NotIncar = len(year10) - num10incarc

year11 = data0912subset[(data0912subset.complaint_year == 2011)]
num11incarc = year11.incar.count()
num11NotIncar = len(year11) - num11incarc

year12 = data0912subset[(data0912subset.complaint_year == 2012)]
num12incarc = year12.incar.count()
num12NotIncar = len(year12) - num12incarc

print "The number of incarcerated people for 2009 is " + str(num09incarc)
print "The number of not incarcerated people for 2009 is " + str(num09NotIncar)
print "The number of total cases for 2009 is " + str(len(year09))
print "\n"

print "The number of incarcerated people for 2010 is " + str(num10incarc)
print "The number of not incarcerated people for 2010 is " + str(num10NotIncar)
print "The number of total cases for 2010 is " + str(len(year10))
print "\n"

print "The number of incarcerated people for 2011 is " + str(num11incarc)
print "The number of not incarcerated people for 2011 is " + str(num11NotIncar)
print "The number of total cases for 2011 is " + str(len(year11))
print "\n"

print "The number of incarcerated people for 2012 is " + str(num12incarc)
print "The number of not incarcerated people for 2012 is " + str(num12NotIncar)
print "The number of total cases for 2012 is " + str(len(year12))

index = ("2009", "2010", "2011", "2012")
df2 = pd.DataFrame({ 
'Incarceration': pd.Series((num09incarc, num10incarc, num11incarc, num12incarc), index),
'No Incarceration': pd.Series((num09NotIncar, num10NotIncar, num11NotIncar, num12NotIncar), index)
})

ax=df2.plot(kind='bar', legend=True, title= "HnR with Highest Charges (of see above)") 
ax.set_ylabel('# of cases')
ax.set_xticklabels(df2.index, rotation='horizontal')

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
avg0912incar = (num09incarc + num10incarc + num11incarc + num12incarc) / 4
avg0912NotIncar = (num09NotIncar + num10NotIncar + num11NotIncar + num12NotIncar) / 4
print "Averaging the 2009-2012 summary, the number of incarcerated people for 2008 can be estimated as " + str(avg0912incar)
print "Averaging the 2009-2012 summary, the number of not incarcerated people for 2008 can be estimated as " + str(avg0912NotIncar)
```
